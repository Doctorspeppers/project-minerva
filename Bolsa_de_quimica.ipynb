usada para substituir as incognitas da equação pelos valores ja

```
import sympy 
from sympy import *
import sympy.parsing.sympy_parser as spp
import pandas as pd
import numpy as np

def achafunc(array, incog):
  tempdict = []
  count = 0
  confirmacao = False
  for equacao in array:
    if len(equacao[2]) == len(incog.values()):
      for x in range(0, len(equacao[2])):
        if equacao[2][x] in incog.values():
          confirmacao = True
          count = count +1
          if count == len(incog.values()):
            print(equacao)
            return equacao[1]
        else:
          confirmacao = False
      if confirmacao == True:
        return equacao[1]
      count = 0
    else:
      pass
  return False
    

def trocavariavel(conta, var):
  conta = conta.split(" ")
  for x in range(0, len(conta)):
    for z in var:
      conta[x] = conta[x].replace(str(z), str(var[z]))

  conta = " ".join(conta)
  conta = conta.split("=")
  return conta

def achaincog(array):

  for i in array.values():
    if i in array.keys():
      return i
  
```
